# Classification Dataset

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/harjotspahwa/Car-Evaluation/refs/heads/master/car_evaluation.csv")
df.columns = ['buying', 'maint', 'doors', 'persons', 'lug_boot', 'safety', 'class']
feature_names = ['buying', 'maint', 'doors', 'persons', 'lug_boot', 'safety']
df
```

```python
X = df.drop("class", axis=1).to_numpy()
y = df["class"].to_numpy()
```

```python
from mlkit.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2, random_state=42)
```

```python
from mlkit.tree import DecisionTreeClassifier

model = DecisionTreeClassifier(criterion="log_loss")
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

```python
model.plot_tree(feature_names=feature_names)
```

```python
np.unique(y)
```

```python
model.predict(X_test)
```

```python
model.predict_proba(X_test)
```

# Regression Dataset

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/likarajo/petrol_consumption/refs/heads/master/data/petrol_consumption.csv")
X = df.drop("Petrol_Consumption", axis=1)
feature_names = list(X.columns)
X = X.to_numpy()
y = df["Petrol_Consumption"].to_numpy()
```

```python
from mlkit.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.3, random_state=42)
```

```python
from mlkit.tree import DecisionTreeRegressor
model = DecisionTreeRegressor(max_depth=10, min_samples_split=5, min_samples_leaf=2, max_leaf_nodes=100)
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

```python
model.predict(X_test)
```

```python
y_test
```

```python
model.plot_tree(feature_names=feature_names)
```
