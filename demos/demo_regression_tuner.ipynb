```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import random
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/Codecademy/datasets/master/streeteasy/streeteasy.csv")
df = pd.get_dummies(df, dtype=int)
df.drop(["rental_id", "building_id"], axis=1, inplace=True)
df.head()
```

```python
df.describe()
```

```python
X = df.drop("rent", axis=1).to_numpy()
y = df["rent"].to_numpy()
```

```python
from mlkit.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.05, random_state=42)
```

```python
from mlkit.tune import ModelTuner
from mlkit.linear_model import LinearRegression, LogisticRegression
from mlkit.model_selection import KFold

tuner = ModelTuner(model = LinearRegression())

best_model = tuner.fit(X_train, y_train)
```

```python
best_model.score(X_test, y_test)
```
